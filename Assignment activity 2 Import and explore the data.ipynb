```python
# Filter the data for the region Gibraltar on "Cases" DataFrame
# Creating a subset with following columns: Deaths, Cases, Recovered and Hospitalised 
# Included also Date column, otherwise data doesnot make sense

gibraltar_cases = cases[cases['Province/State'] == 'Gibraltar']

gibraltar_cases = gibraltar_cases[['Date', 'Deaths', 'Cases', 'Recovered', 'Hospitalised']]


# Exploring "gibraltar_cases" DataFrame
print(gibraltar_cases.shape)

gibraltar_cases_na = gibraltar_cases[gibraltar_cases.isnull().any(axis = 1)]
print(gibraltar_cases_na.shape)     # Checking for any missing values

```
