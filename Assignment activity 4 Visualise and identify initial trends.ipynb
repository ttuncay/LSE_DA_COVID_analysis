Assignment activity 3: Merge and analyse the data

Objective:
Identify the area that has the greatest number of people who have received a first dose but not a second dose.

```python
# Merging 2 DataFrames cases & vaccinated 

uk_case_and_vaccine = cases.merge(vaccinated, how = 'left', on = ['Lat', 'Long', 'Date'])
```

```python
# Checking merge result, checking for missing values and identifying columns etc.

print(uk_case_and_vaccine.shape)     #check for the size of Df
print(uk_case_and_vaccine[uk_case_and_vaccine.isna().any(axis = 1)].shape)     #Check for missing values
uk_case_and_vaccine.head()
```

```python
# Selecting necessary columns to keep, dropping unnecessary ones
uk_case_and_vaccine = uk_case_and_vaccine.drop(['Province/State_y', 
                                                'Country/Region_y', 
                                                'ISO 3166-1 Alpha 3-Codes_y', 
                                                'Sub-region Name_y', 
                                                'Intermediate Region Code_y'], axis=1)
```

```python
# Viewing df for crosscheck of action
uk_case_and_vaccine.head()
```

```python
# Renaming column names
uk_case_and_vaccine.columns = ['Province/State', 
                               'Country/Region', 
                               'Lat', 
                               'Long',
                               'ISO 3166-1 Alpha 3-Codes',
                               'Sub-region Name',
                               'Intermediate Region Code',
                               'Date',
                               'Deaths',
                               'Cases',
                               'Recovered',
                               'Hospitalised',
                               'Vaccinated',
                               'First Dose',
                               'Second Dose']
```

```python
# Viewing df for crosscheck of action
uk_case_and_vaccine.head()
```

```python
# Viewing dtypes of merged dataframe
print(uk_case_and_vaccine.dtypes)
```

```python
# Importing DateTime library 
import datetime
from datetime import datetime, date
```

```python
# Converting datatype of Date column from object to DateTime
uk_case_and_vaccine['Date'] = pd.to_datetime(uk_case_and_vaccine['Date'])

# Viewing dtypes of merged dataframe again
print(uk_case_and_vaccine.dtypes)
```

```python
# Grouping and creating new df under province name showing sum of vaccined, first dose and second dose
vaccine_province = uk_case_and_vaccine.groupby('Province/State')['Vaccinated', 'First Dose', 'Second Dose'].sum()
```

```python
# Calculating number of people who have received a first dose but not a second dose.
vaccine_province['Delta'] = vaccine_province['First Dose']-vaccine_province['Second Dose']

# Viewing df for crosscheck of action
vaccine_province.head()
```

```python
# Which Province/State has the highest number of individuals who have received a first dose but not a second dose?
vaccine_province.sort_values('Delta', ascending=False)
```

```python
# Which Province/State has the highest percentage of individuals who have received a first dose but not a second dose.

# Calculating percentage of people who received the first dose but not the second
vaccine_province['not fully vaccinated %'] = (vaccine_province['Delta'] / vaccine_province['First Dose'])*100
vaccine_province.sort_values('not fully vaccinated %', ascending=False)
```

```python
# Grouping and creating new df under date showing sum of vaccined, first dose and second dose
vaccine_date = uk_case_and_vaccine.groupby('Date')['Vaccinated', 'First Dose', 'Second Dose'].sum()
vaccine_date
```

```python
# Exloring further dates and trying to understand why having peaks on some days. 

# Moving index(date) to column 
vaccine_date.reset_index(inplace = True)
vaccine_date.head()
```

```python
# Bring day names as a new columnn
vaccine_date['Weekday'] = vaccine_date['Date'].apply(lambda x : x.strftime('%A'))
```

```python
# Reordering the column names for easy interpretation
vaccine_date = vaccine_date[['Date', 'Weekday', 'Vaccinated', 'First Dose', 'Second Dose']]
vaccine_date
```

```python
# Import Matplotlib, Seaborn, NumPy, and Pandas.
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import datetime
```

```python
# Create a chart that communicates the number of individuals who have received a second dose 
# compared to individuals who have received their first dose in each area.
# Calling back vaccine_province df to work on

# Renaming columns according to guadinace
vaccine_province = vaccine_province.rename(columns={'not fully vaccinated %':'ratio of interest %','Delta':'difference per region'})

#Creating a new column thats shows percantage of people fully vaccinated (2 doses)
vaccine_province['fully vaccinated %']=vaccine_province['Second Dose']/vaccine_province['First Dose']*100

# Bringing province data as a column
vaccine_province = vaccine_province.reset_index()
vaccine_province
```

```python
# Preparing data for barplot using melt function to create long df
# Barplot  showing first dose and second dose takers per province

vaccine_province_barplot1 = vaccine_province.melt(id_vars='Province/State', value_vars=['First Dose','Second Dose'])

# Renaming columns
vaccine_province_barplot1 = vaccine_province_barplot1.rename(columns={'variable':'Vaccine Dose', 'value':'Number of People'})

#Checking types to make sure I have catagorical value for x axis
print(vaccine_province_barplot1.dtypes) 
vaccine_province_barplot1     
```

```python
# Barplot representation

# Adjusting the size and adding title
plt.figure(figsize=(15,8))
plt.title('First Dose vs Second Dose per Province/State')

# Creating the bar plot
ax = sns.barplot(x='Province/State',y='Number of People', hue='Vaccine Dose', estimator=sum, data=vaccine_province_barplot1)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)     #Rotating x labels for easy readibility
```

```python
#Saving the bar plot
fig = ax.get_figure()
fig.patch.set_facecolor('white')
fig.savefig('Vaccine_Province_Barplot1.png')
```

```python
# Preparing data for barplot using melt function to create long df
# Bar plot of % fully vaccinated vs ratio of interest (not fully vaccinated) per province/region

vaccine_province_barplot2 = vaccine_province.melt(id_vars='Province/State', value_vars=['fully vaccinated %','ratio of interest %'])

# Renaming columns
vaccine_province_barplot2 = vaccine_province_barplot2.rename(columns={'variable':'full vaccinated or not', 'value':'% of People'})

#Checking types to make sure I have catagorical value for x axis
print(vaccine_province_barplot2.dtypes) 
vaccine_province_barplot2     
```

```python
# Barplot representation

# Adjusting the size and adding title
plt.figure(figsize=(15,8))
plt.title('Fully Vaccinated % vs Ratio of Interest% per Province/State')

# Creating the bar plot
ax = sns.barplot(x='Province/State',y='% of People', hue='full vaccinated or not', estimator=sum, data=vaccine_province_barplot2)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)     #Rotating x labels for easy readibility
```

```python
#Saving the bar plot
fig = ax.get_figure()
fig.patch.set_facecolor('white')
fig.savefig('Vaccine_Province_Barplot2.png')
```

```python
#Create a visualisation that communicates the number of deaths per month in each region. 
#Review the visualisation to determine whether there are areas that hit higher peaks earlier than other areas.

#Group the data by Province/State and Date, and aggregate the death count.
death_province_lineplot1=uk_case_and_vaccine.groupby(['Province/State','Date'])[['Deaths']].sum()
death_province_lineplot1
```

```python
# Create a lineplot to display the trend of deaths across all regions over time. 

#Bringing province and date data to columns
death_province_lineplot1=death_province_lineplot1.reset_index()
print(death_province_lineplot1.dtypes)

# Adjusting the size and adding title
plt.figure(figsize=(15,8))
plt.title('Deaths over time per Province/State')

# Creating the line plot
ax = sns.lineplot(x='Date',y='Deaths', hue='Province/State', data=death_province_lineplot1)
```

```python
# Adjusting the size and adding title
plt.figure(figsize=(15,8))
plt.title('Deaths over time per Province/State')

# ReCreating the line plot without Others that distort the data
ax = sns.lineplot(x='Date',y='Deaths', hue='Province/State', data=death_province_lineplot1[death_province_lineplot1['Province/State']!='Others'], palette='colorblind')
```

```python
#Saving the line plot
fig = ax.get_figure()
fig.patch.set_facecolor('white')
fig.savefig('Death_Province_Lineplot1.png')
```

```python
#Convert Date into Months and plot the same line graph as previously.

#Creating new dataframe for this visual
death_province_lineplot2=death_province_lineplot1.copy()
death_province_lineplot2['Month']=pd.to_datetime(death_province_lineplot2['Date']+pd.offsets.MonthBegin(0))

# Adjusting the size and adding title
plt.figure(figsize=(15,8))
plt.title('Deaths over time per Province/State')

# ReCreating the line plot without Others that distort the data
ax = sns.lineplot(x='Month',y='Deaths', hue='Province/State', data=death_province_lineplot2[death_province_lineplot2['Province/State']!='Others'], palette='colorblind')
```

```python
#Saving the line plot
fig = ax.get_figure()
fig.patch.set_facecolor('white')
fig.savefig('Death_Province_Lineplot2.png')
```

```python
#Identify which region has the greatest number of recoveries. 
#Create a visualisation that communicates the number of recoveries per month in each area. 
#Review the visualisation to determine whether this has been consistent over time.

#Group the data by Province/State, and aggregate the count of recovered cases
#Sort the values of recovered cases in ascending order.

recovered_province_lineplot=uk_case_and_vaccine.groupby(['Province/State','Date'])[['Recovered']].sum()
recovered_province_lineplot=recovered_province_lineplot.reset_index()

#Converting Date to Months
recovered_province_lineplot['Month']=pd.to_datetime(recovered_province_lineplot['Date']+pd.offsets.MonthBegin(0))

print(recovered_province_lineplot.dtypes)
recovered_province_lineplot.head()
```

```python
# Adjusting the size and adding title
plt.figure(figsize=(15,8))
plt.title('Recovered over time per Province/State')

# ReCreating the line plot without Others that distort the data
ax = sns.lineplot(x='Month',y='Recovered', hue='Province/State', data=recovered_province_lineplot, palette='colorblind')
```

```python
#Saving the recovered plot
fig = ax.get_figure()
fig.patch.set_facecolor('white')
fig.savefig('Death_Province_Lineplot2.png')
```
